```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=42, hypercube=False, class_sep=10)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
```

```python
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])

    epochs = 1000
    learning_rate = 0.1
    for i in range(epochs):
        j = np.random.randint(0, 100)
        y_pred = step_function(np.dot(X[j], weights))
        weights = weights + learning_rate * (y[j] - y_pred) * X[j]

    return weights[0], weights[1:]
```

```python
def step_function(z):
    return 1 if z >= 0 else -1
```

```python
intercept_, coef_ = perceptron(X, y)
```

```python
print(intercept_)
print(coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=50)
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.show()
```
